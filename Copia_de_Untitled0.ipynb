```python
cantidad = 1000000
```

```python
import random
```

```python
#Datos aleatorios
nombres = ["Juan", "Sofía", "Miguel", "Laura", "Carlos", "Ana", "José", "Isabel"]
ciudades = ["Madrid", "Barcelona", "Sevilla", "Valencia", "Bilbao", "Málaga", "Zaragoza", "Murcia"]
categorias = ["Electrónica", "Ropa", "Hogar", "Libros", "Deportes", "Alimentos", "Juguetes", "Belleza"]
```

```python
#Creación del DataFrame
dfMillon = pd.DataFrame ({
    "nombre" : [random.choice(nombres) for _ in range(cantidad)],
    "edad" : [random.randint(18,70) for _ in range(cantidad) ],
    "salario" : [round(random.uniform(5000,50000),2) for _ in range(cantidad)],
    "activo" : [random.choice([True, False]) for _ in range(cantidad)],
    "categoria" : [random.choice(ciudades) for _ in range(cantidad)],
    "anio_registro" : [random.randint(2020,2026) for _ in range(cantidad)]
})
```
